```python
# deleting limiter or marked in every start of the page
data_blib = data_blib.iloc[0:-6]
def delete_mark(data):
  for i in range (len(data)):
    data[i] = re.sub(r'[\ ]+(VALID_DATA_BELOW)$', '', data[i])
    data[i] = re.sub(r'[\.]', '', data[i])

delete_mark(data_blib['text'])
data_blib['text'] = data_blib['text'].str.lower()
data_blib
```

```python
tag_sort_rele = 'blibli sort by relevance'
tag_sort_lowe = 'blibli sort by lowest price'
tag_sort_high = 'blibli sort by highest price'
tag_filt_lowe = 'blibli filter by lowest price'
tag_filt_high = 'blibli filter by highest price'

def find_class(data, detail_data, marker):
  temp_list = list( map(int, data.index[(detail_data == marker)].tolist()) )
  return temp_list

tag_sort_rele_num = find_class(data_blib, data_blib['text'], tag_sort_rele)
tag_sort_lowe_num = find_class(data_blib, data_blib['text'], tag_sort_lowe)
tag_sort_high_num = find_class(data_blib, data_blib['text'], tag_sort_high)
tag_filt_lowe_num = find_class(data_blib, data_blib['text'], tag_filt_lowe)
tag_filt_high_num = find_class(data_blib, data_blib['text'], tag_filt_high)

tag_sort_rele_num = tag_sort_rele_num[0]
tag_sort_lowe_num = tag_sort_lowe_num[0]
tag_sort_high_num = tag_sort_high_num[0]
tag_filt_lowe_num = tag_filt_lowe_num[0]
tag_filt_high_num = tag_filt_high_num[0]

blib_sort_rele    =   data_blib.iloc[tag_sort_rele_num+1:tag_sort_lowe_num]
blib_sort_lowe    =   data_blib.iloc[tag_sort_lowe_num+1:tag_sort_high_num]
blib_sort_high    =   data_blib.iloc[tag_sort_high_num+1:tag_filt_lowe_num]
blib_filt_lowe    =   data_blib.iloc[tag_filt_lowe_num+1:tag_filt_high_num]
blib_filt_high    =   data_blib.iloc[tag_filt_high_num+1:]

blib_sort_rele.reset_index(inplace=True, drop=True)
blib_sort_lowe.reset_index(inplace=True, drop=True)
blib_sort_high.reset_index(inplace=True, drop=True)
blib_filt_lowe.reset_index(inplace=True, drop=True)
blib_filt_high.reset_index(inplace=True, drop=True)
```

## cleaning data

```python
def regexing_word(data_details):
  for i in range (len(data_details)):
    data_details[i] = re.sub(r'^[\(]*[\d]{1,2}[\)]*$', 'deleted', data_details[i])           # rating tag
    data_details[i] = re.sub(r'(terjual) [0-9+]+[ ]*[rb+]*', 'deleted', data_details[i])                    # terjual tag
    data_details[i] = re.sub(r'(-){78}', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(grosir)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^\d\,\d$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^\d\%\(rp\d+\)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(voucher seller)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(pre\-order)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(tambah ke bag)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(hingga )[rp]*\d+\%*$', 'deleted', data_details[i])	
    data_details[i] = re.sub(r'^(promo sd)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^\d{0,2}( ags 2022)$', 'deleted', data_details[i])	
    data_details[i] = re.sub(r'^(anak)$', 'deleted', data_details[i])	
    data_details[i] = re.sub(r'^(coba tambahkan kata pencarian yang lebih spesifik\, seperti\:)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(valid_data_below)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(lihat detail)$', 'deleted', data_details[i])	

regexing_word(blib_sort_rele['text'])
regexing_word(blib_sort_lowe['text'])
regexing_word(blib_sort_high['text'])
regexing_word(blib_filt_lowe['text'])
regexing_word(blib_filt_high['text'])

blib_sort_rele = blib_sort_rele[blib_sort_rele['text'].str.contains("deleted")==False]
blib_sort_lowe = blib_sort_lowe[blib_sort_lowe['text'].str.contains("deleted")==False]
blib_sort_high = blib_sort_high[blib_sort_high['text'].str.contains("deleted")==False]
blib_filt_lowe = blib_filt_lowe[blib_filt_lowe['text'].str.contains("deleted")==False]
blib_filt_high = blib_filt_high[blib_filt_high['text'].str.contains("deleted")==False]

blib_sort_rele.reset_index(inplace=True, drop=True)
blib_sort_lowe.reset_index(inplace=True, drop=True)
blib_sort_high.reset_index(inplace=True, drop=True)
blib_filt_lowe.reset_index(inplace=True, drop=True)
blib_filt_high.reset_index(inplace=True, drop=True)
```

```python
keyw_1 = '%'
keyw_2 = 'rp'
keyw_3 = '| pass |'


def removing_word(data_details):
  for i in range (len(data_details)):
    if keyw_2 in data_details[i]:
      if keyw_2 in data_details[i+1]:
        if keyw_1 in data_details[i+2]:
          data_details[i+1] = 'deleted'
          data_details[i+2] = 'deleted'
      else:
        continue
    elif keyw_3 in data_details[i]:
      data_details[i] = 'deleted'
    else:
      continue

removing_word(blib_sort_rele['text'])
removing_word(blib_sort_lowe['text'])
removing_word(blib_sort_high['text'])
removing_word(blib_filt_lowe['text'])
removing_word(blib_filt_high['text'])

blib_sort_rele = blib_sort_rele[blib_sort_rele['text'].str.contains("deleted")==False]
blib_sort_lowe = blib_sort_lowe[blib_sort_lowe['text'].str.contains("deleted")==False]
blib_sort_high = blib_sort_high[blib_sort_high['text'].str.contains("deleted")==False]
blib_filt_lowe = blib_filt_lowe[blib_filt_lowe['text'].str.contains("deleted")==False]
blib_filt_high = blib_filt_high[blib_filt_high['text'].str.contains("deleted")==False]

blib_sort_rele.reset_index(inplace=True, drop=True)
blib_sort_lowe.reset_index(inplace=True, drop=True)
blib_sort_high.reset_index(inplace=True, drop=True)
blib_filt_lowe.reset_index(inplace=True, drop=True)
blib_filt_high.reset_index(inplace=True, drop=True)
```
