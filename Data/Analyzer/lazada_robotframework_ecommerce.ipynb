```python
# deleting limiter or marked in every start of the page
data_laza = data_laza.iloc[0:-6]
def delete_mark(data):
  for i in range (len(data)):
    data[i] = re.sub(r'[\ ]+(VALID_DATA_BELOW)$', '', data[i])
    data[i] = re.sub(r'[\.]', '', data[i])

delete_mark(data_laza['text'])
data_laza['text'] = data_laza['text'].str.lower()
```

```python
tag_sort_rele = 'lazada sort by relevance'
tag_sort_lowe = 'lazada sort by lowest price'
tag_sort_high = 'lazada sort by highest price'
tag_filt_lowe = 'lazada filter by lowest price'
tag_filt_high = 'lazada filter by highest price'

def find_class(data, detail_data, marker):
  temp_list = list( map(int, data.index[(detail_data == marker)].tolist()) )
  return temp_list

tag_sort_rele_num = find_class(data_laza, data_laza['text'], tag_sort_rele)
tag_sort_lowe_num = find_class(data_laza, data_laza['text'], tag_sort_lowe)
tag_sort_high_num = find_class(data_laza, data_laza['text'], tag_sort_high)
tag_filt_lowe_num = find_class(data_laza, data_laza['text'], tag_filt_lowe)
tag_filt_high_num = find_class(data_laza, data_laza['text'], tag_filt_high)

tag_sort_rele_num = tag_sort_rele_num[0]
tag_sort_lowe_num = tag_sort_lowe_num[0]
tag_sort_high_num = tag_sort_high_num[0]
tag_filt_lowe_num = tag_filt_lowe_num[0]
tag_filt_high_num = tag_filt_high_num[0]

laza_sort_rele    =   data_laza.iloc[tag_sort_rele_num+1:tag_sort_lowe_num]
laza_sort_lowe    =   data_laza.iloc[tag_sort_lowe_num+1:tag_sort_high_num]
laza_sort_high    =   data_laza.iloc[tag_sort_high_num+1:tag_filt_lowe_num]
laza_filt_lowe    =   data_laza.iloc[tag_filt_lowe_num+1:tag_filt_high_num]
laza_filt_high    =   data_laza.iloc[tag_filt_high_num+1:]

laza_sort_rele.reset_index(inplace=True, drop=True)
laza_sort_lowe.reset_index(inplace=True, drop=True)
laza_sort_high.reset_index(inplace=True, drop=True)
laza_filt_lowe.reset_index(inplace=True, drop=True)
laza_filt_high.reset_index(inplace=True, drop=True)
```

## cleaning data

```python
def regexing_word(data_details):
  for i in range (len(data_details)):
    data_details[i] = re.sub(r'^\(\d+\)$', 'deleted', data_details[i])           # rating tag=
    data_details[i] = re.sub(r'(-){78}', 'deleted', data_details[i])  
    data_details[i] = re.sub(r'^\d+\%(off)$', 'deleted', data_details[i]) 
    data_details[i] = re.sub(r'^(murah)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(flash sale\, berakhir)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(flash sale)$', 'deleted', data_details[i])  
    data_details[i] = re.sub(r'^(baru)$', 'deleted', data_details[i])   
    data_details[i] = re.sub(r'^[\-]*\d{0,2}\d+\%$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(valid_data_below)$', 'deleted', data_details[i])

regexing_word(laza_sort_rele['text'])
regexing_word(laza_sort_lowe['text'])
regexing_word(laza_sort_high['text'])
regexing_word(laza_filt_lowe['text'])
regexing_word(laza_filt_high['text'])

laza_sort_rele = laza_sort_rele[laza_sort_rele['text'].str.contains("deleted")==False]
laza_sort_lowe = laza_sort_lowe[laza_sort_lowe['text'].str.contains("deleted")==False]
laza_sort_high = laza_sort_high[laza_sort_high['text'].str.contains("deleted")==False]
laza_filt_lowe = laza_filt_lowe[laza_filt_lowe['text'].str.contains("deleted")==False]
laza_filt_high = laza_filt_high[laza_filt_high['text'].str.contains("deleted")==False]

laza_sort_rele.reset_index(inplace=True, drop=True)
laza_sort_lowe.reset_index(inplace=True, drop=True)
laza_sort_high.reset_index(inplace=True, drop=True)
laza_filt_lowe.reset_index(inplace=True, drop=True)
laza_filt_high.reset_index(inplace=True, drop=True)
```

```python
keyw_1 = 'rp'
keyw_2 = '| pass |'


def removing_word(data_details):
  for i in range (len(data_details)):
    if re.match('^rp\d+$', data_details[i]):
      if re.match('^rp\d+$', data_details[i+1]):
        data_details[i+1] = 'deleted'
      else:
        continue
    elif keyw_2 in data_details[i]:
      data_details[i] = 'deleted'
    else:
      continue

removing_word(laza_sort_rele['text'])
removing_word(laza_sort_lowe['text'])
removing_word(laza_sort_high['text'])
removing_word(laza_filt_lowe['text'])
removing_word(laza_filt_high['text'])

laza_sort_rele = laza_sort_rele[laza_sort_rele['text'].str.contains("deleted")==False]
laza_sort_lowe = laza_sort_lowe[laza_sort_lowe['text'].str.contains("deleted")==False]
laza_sort_high = laza_sort_high[laza_sort_high['text'].str.contains("deleted")==False]
laza_filt_lowe = laza_filt_lowe[laza_filt_lowe['text'].str.contains("deleted")==False]
laza_filt_high = laza_filt_high[laza_filt_high['text'].str.contains("deleted")==False]

laza_sort_rele.reset_index(inplace=True, drop=True)
laza_sort_lowe.reset_index(inplace=True, drop=True)
laza_sort_high.reset_index(inplace=True, drop=True)
laza_filt_lowe.reset_index(inplace=True, drop=True)
laza_filt_high.reset_index(inplace=True, drop=True)
```

```python
elec_1 = ('case')
elec_2 = ('redmi')
elec_3 = ('10a')
elec_4 = ('10 a')

def elec_chec_keyw(data_details):
  falseCounter=0
  for i in range (len(data_details)):
    if elec_1 in data_details[i]:
      if elec_2 in data_details[i]:
        if elec_3 in data_details[i]:
          continue
        if elec_4 in data_details[i]:
          continue
        else:
          falseCounter+=1
      else:
          falseCounter+=1
    else:
      falseCounter+=1
  print('accuracy of sort by relevance lazada: ', (len(data_details)-falseCounter)/len(data_details))
  print('true items, :', len(data_details)-falseCounter, ', false items :', falseCounter)
```
