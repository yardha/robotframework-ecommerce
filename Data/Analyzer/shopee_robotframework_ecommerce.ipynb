```python
# deleting limiter or marked in every start of the page
data_shop = data_shop.iloc[0:-6]
def delete_mark(data):
  for i in range (len(data)):
    data[i] = re.sub(r'[\ ]+(VALID_DATA_BELOW)$', '', data[i])
    data[i] = re.sub(r'[\.]', '', data[i])

delete_mark(data_shop['text'])
data_shop['text'] = data_shop['text'].str.lower()
```

```python
tag_sort_rele = 'shopee sort by relevance'
tag_sort_lowe = 'shopee sort by lowest price'
tag_sort_high = 'shopee sort by highest price'
tag_filt_lowe = 'shopee filter by lowest price'
tag_filt_high = 'shopee filter by highest price'

def find_class(data, detail_data, marker):
  temp_list = list( map(int, data.index[(detail_data == marker)].tolist()) )
  return temp_list

tag_sort_rele_num = find_class(data_shop, data_shop['text'], tag_sort_rele)
tag_sort_lowe_num = find_class(data_shop, data_shop['text'], tag_sort_lowe)
tag_sort_high_num = find_class(data_shop, data_shop['text'], tag_sort_high)
tag_filt_lowe_num = find_class(data_shop, data_shop['text'], tag_filt_lowe)
tag_filt_high_num = find_class(data_shop, data_shop['text'], tag_filt_high)

tag_sort_rele_num = tag_sort_rele_num[0]
tag_sort_lowe_num = tag_sort_lowe_num[0]
tag_sort_high_num = tag_sort_high_num[0]
tag_filt_lowe_num = tag_filt_lowe_num[0]
tag_filt_high_num = tag_filt_high_num[0]

shop_sort_rele    =   data_shop.iloc[tag_sort_rele_num+1:tag_sort_lowe_num]
shop_sort_lowe    =   data_shop.iloc[tag_sort_lowe_num+1:tag_sort_high_num]
shop_sort_high    =   data_shop.iloc[tag_sort_high_num+1:tag_filt_lowe_num]
shop_filt_lowe    =   data_shop.iloc[tag_filt_lowe_num+1:tag_filt_high_num]
shop_filt_high    =   data_shop.iloc[tag_filt_high_num+1:]

shop_sort_rele.reset_index(inplace=True, drop=True)
shop_sort_lowe.reset_index(inplace=True, drop=True)
shop_sort_high.reset_index(inplace=True, drop=True)
shop_filt_lowe.reset_index(inplace=True, drop=True)
shop_filt_high.reset_index(inplace=True, drop=True)
```

```python
def regexing_word(data_details):
  for i in range (len(data_details)):
    data_details[i] = re.sub(r'^iklan$', 'deleted', data_details[i])                      # ad tag
    data_details[i] = re.sub(r'^star$', 'deleted', data_details[i])           # rating tag
    data_details[i] = re.sub(r'(^[\d]+[,]*[\d]*[ rb+]*( terjual)+)', 'deleted', data_details[i])                    # terjual tag
    data_details[i] = re.sub(r'^off$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^\d{0,2}$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(-){78}', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(grosir)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(produk serupa)$', 'deleted', data_details[i])                    # terjual tag
    data_details[i] = re.sub(r'^(kombo hemat)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(murah lebay)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(hadiah gratis)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(diskon)[ ]+[rp \d\%\.]+', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^\d{1,2}\%$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(cashback)[ ]\d+\%$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(valid_data_below)$', 'deleted', data_details[i])

regexing_word(shop_sort_rele['text'])
regexing_word(shop_sort_lowe['text'])
regexing_word(shop_sort_high['text'])
regexing_word(shop_filt_lowe['text'])
regexing_word(shop_filt_high['text'])

shop_sort_rele = shop_sort_rele[shop_sort_rele['text'].str.contains("deleted")==False]
shop_sort_lowe = shop_sort_lowe[shop_sort_lowe['text'].str.contains("deleted")==False]
shop_sort_high = shop_sort_high[shop_sort_high['text'].str.contains("deleted")==False]
shop_filt_lowe = shop_filt_lowe[shop_filt_lowe['text'].str.contains("deleted")==False]
shop_filt_high = shop_filt_high[shop_filt_high['text'].str.contains("deleted")==False]

shop_sort_rele.reset_index(inplace=True, drop=True)
shop_sort_lowe.reset_index(inplace=True, drop=True)
shop_sort_high.reset_index(inplace=True, drop=True)
shop_filt_lowe.reset_index(inplace=True, drop=True)
shop_filt_high.reset_index(inplace=True, drop=True)
```

```python
keyw_1 = 'rp'
keyw_2 = '%'
keyw_3 = '| pass |'


def removing_word(data_details):
  for i in range (len(data_details)):
    if re.match('^rp\d+$', data_details[i]):
      if re.match('^rp\d+$', data_details[i+1]):
          data_details[i] = 'deleted'
      else:
        continue
    elif keyw_3 in data_details[i]:
      data_details[i] = 'deleted'
    elif re.match('(pilih )\d', data_details[i]):
      data_details[i] = 'deleted'
    else:
      continue

removing_word(shop_sort_rele['text'])
removing_word(shop_sort_lowe['text'])
removing_word(shop_sort_high['text'])
removing_word(shop_filt_lowe['text'])
removing_word(shop_filt_high['text'])

shop_sort_rele = shop_sort_rele[shop_sort_rele['text'].str.contains("deleted")==False]
shop_sort_lowe = shop_sort_lowe[shop_sort_lowe['text'].str.contains("deleted")==False]
shop_sort_high = shop_sort_high[shop_sort_high['text'].str.contains("deleted")==False]
shop_filt_lowe = shop_filt_lowe[shop_filt_lowe['text'].str.contains("deleted")==False]
shop_filt_high = shop_filt_high[shop_filt_high['text'].str.contains("deleted")==False]

shop_sort_rele.reset_index(inplace=True, drop=True)
shop_sort_lowe.reset_index(inplace=True, drop=True)
shop_sort_high.reset_index(inplace=True, drop=True)
shop_filt_lowe.reset_index(inplace=True, drop=True)
shop_filt_high.reset_index(inplace=True, drop=True)
```

```python
keyw_1 = 'rp'

def remove_unused_words(data_details):
  for i in range (len(data_details)-2):
    if re.match('^rp\d+( - rp)*\d*$', data_details[i]):
      if re.match('^rp\d+( - rp)*\d*$', data_details[i+2]):
        continue
      else:
        data_details[i+1] = 'deleted'
    elif re.match('^(diskon )(rp)\d(rb)',data_details[i]):
      data_details[i] = 'deleted'
    else:
      continue
  data_details[len(data_details)-1] = 'deleted'

remove_unused_words(shop_sort_rele['text'])
remove_unused_words(shop_sort_lowe['text'])
remove_unused_words(shop_sort_high['text'])
remove_unused_words(shop_filt_lowe['text'])
remove_unused_words(shop_filt_high['text'])

shop_sort_rele = shop_sort_rele[shop_sort_rele['text'].str.contains("deleted")==False]
shop_sort_lowe = shop_sort_lowe[shop_sort_lowe['text'].str.contains("deleted")==False]
shop_sort_high = shop_sort_high[shop_sort_high['text'].str.contains("deleted")==False]
shop_filt_lowe = shop_filt_lowe[shop_filt_lowe['text'].str.contains("deleted")==False]
shop_filt_high = shop_filt_high[shop_filt_high['text'].str.contains("deleted")==False]

shop_sort_rele.reset_index(inplace=True, drop=True)
shop_sort_lowe.reset_index(inplace=True, drop=True)
shop_sort_high.reset_index(inplace=True, drop=True)
shop_filt_lowe.reset_index(inplace=True, drop=True)
shop_filt_high.reset_index(inplace=True, drop=True)
```
