```python
# deleting limiter or marked in every start of the page
data_toko = data_toko.iloc[0:-6]
def delete_mark(data):
  for i in range (len(data)):
    data[i] = re.sub(r'[\ ]+(VALID_DATA_BELOW)$', '', data[i])
    data[i] = re.sub(r'[\.]', '', data[i])

delete_mark(data_toko['text'])
data_toko['text'] = data_toko['text'].str.lower()
```

```python
tag_sort_rele = 'tokopedia sort by relevance'
tag_sort_lowe = 'tokopedia sort by lowest price'
tag_sort_high = 'tokopedia sort by highest price'
tag_filt_lowe = 'tokopedia filter by lowest price'
tag_filt_high = 'tokopedia filter by highest price'

def find_class(data, detail_data, marker):
  temp_list = list( map(int, data.index[(detail_data == marker)].tolist()) )
  return temp_list

tag_sort_rele_num = find_class(data_toko, data_toko['text'], tag_sort_rele)
tag_sort_lowe_num = find_class(data_toko, data_toko['text'], tag_sort_lowe)
tag_sort_high_num = find_class(data_toko, data_toko['text'], tag_sort_high)
tag_filt_lowe_num = find_class(data_toko, data_toko['text'], tag_filt_lowe)
tag_filt_high_num = find_class(data_toko, data_toko['text'], tag_filt_high)

tag_sort_rele_num = tag_sort_rele_num[0]
tag_sort_lowe_num = tag_sort_lowe_num[0]
tag_sort_high_num = tag_sort_high_num[0]
tag_filt_lowe_num = tag_filt_lowe_num[0]
tag_filt_high_num = tag_filt_high_num[0]

toko_sort_rele    =   data_toko.iloc[tag_sort_rele_num+1:tag_sort_lowe_num]
toko_sort_lowe    =   data_toko.iloc[tag_sort_lowe_num+1:tag_sort_high_num]
toko_sort_high    =   data_toko.iloc[tag_sort_high_num+1:tag_filt_lowe_num]
toko_filt_lowe    =   data_toko.iloc[tag_filt_lowe_num+1:tag_filt_high_num]
toko_filt_high    =   data_toko.iloc[tag_filt_high_num+1:]

toko_sort_rele.reset_index(inplace=True, drop=True)
toko_sort_lowe.reset_index(inplace=True, drop=True)
toko_sort_high.reset_index(inplace=True, drop=True)
toko_filt_lowe.reset_index(inplace=True, drop=True)
toko_filt_high.reset_index(inplace=True, drop=True)
```

## cleaning data

```python
def regexing_word(data_details):
  for i in range (len(data_details)):
    data_details[i] = re.sub(r'^ad$', 'deleted', data_details[i])                      # ad tag
    data_details[i] = re.sub(r'^\d\d$', 'deleted', data_details[i])           # rating tag
    data_details[i] = re.sub(r'(terjual) [0-9+]+[ ]*[rb+]*', 'deleted', data_details[i])                    # terjual tag
    data_details[i] = re.sub(r'(sisa) [0-9+]+', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(-){78}', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(grosir)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(produk terbaru)$', 'deleted', data_details[i])                    # terjual tag
    data_details[i] = re.sub(r'^(cashback)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(preorder)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(terlaris)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(tokopedia tekno)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(super gadget day)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(galeri fashion)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(lagi trending)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(cantik fest)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(women in style)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(selalu os)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(sportacular)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(ktp medan)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(tokopedia garage)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(tokopediaparents)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'^(home living sale)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(valid_data_below)$', 'deleted', data_details[i])
    data_details[i] = re.sub(r'(pasti surabaya)$', 'deleted', data_details[i])

regexing_word(toko_sort_rele['text'])
regexing_word(toko_sort_lowe['text'])
regexing_word(toko_sort_high['text'])
regexing_word(toko_filt_lowe['text'])
regexing_word(toko_filt_high['text'])

toko_sort_rele = toko_sort_rele[toko_sort_rele['text'].str.contains("deleted")==False]
toko_sort_lowe = toko_sort_lowe[toko_sort_lowe['text'].str.contains("deleted")==False]
toko_sort_high = toko_sort_high[toko_sort_high['text'].str.contains("deleted")==False]
toko_filt_lowe = toko_filt_lowe[toko_filt_lowe['text'].str.contains("deleted")==False]
toko_filt_high = toko_filt_high[toko_filt_high['text'].str.contains("deleted")==False]

toko_sort_rele.reset_index(inplace=True, drop=True)
toko_sort_lowe.reset_index(inplace=True, drop=True)
toko_sort_high.reset_index(inplace=True, drop=True)
toko_filt_lowe.reset_index(inplace=True, drop=True)
toko_filt_high.reset_index(inplace=True, drop=True)
```

```python
keyw_1 = '%'
keyw_2 = 'rp'
keyw_3 = '| pass |'


def removing_word(data_details):
  for i in range (len(data_details)):
    if keyw_1 in data_details[i]:
      if keyw_2 in data_details[i-1]:
        if keyw_2 in data_details[i+1]:
          data_details[i] = 'deleted'
          data_details[i+1] = 'deleted' 
        else:
          continue
      else:
        continue
    elif keyw_3 in data_details[i]:
      data_details[i] = 'deleted'
    else:
      continue

removing_word(toko_sort_rele['text'])
removing_word(toko_sort_lowe['text'])
removing_word(toko_sort_high['text'])
removing_word(toko_filt_lowe['text'])
removing_word(toko_filt_high['text'])

toko_sort_rele = toko_sort_rele[toko_sort_rele['text'].str.contains("deleted")==False]
toko_sort_lowe = toko_sort_lowe[toko_sort_lowe['text'].str.contains("deleted")==False]
toko_sort_high = toko_sort_high[toko_sort_high['text'].str.contains("deleted")==False]
toko_filt_lowe = toko_filt_lowe[toko_filt_lowe['text'].str.contains("deleted")==False]
toko_filt_high = toko_filt_high[toko_filt_high['text'].str.contains("deleted")==False]

toko_sort_rele.reset_index(inplace=True, drop=True)
toko_sort_lowe.reset_index(inplace=True, drop=True)
toko_sort_high.reset_index(inplace=True, drop=True)
toko_filt_lowe.reset_index(inplace=True, drop=True)
toko_filt_high.reset_index(inplace=True, drop=True)
```

```python
keyw_1 = 'rp'

def remove_unused_words(data_details):
  for i in range (len(data_details)):
    if re.match('^rp\d+$', data_details[i]):
      data_details[i+1] = 'deleted'
      data_details[i+2] = 'deleted'
    else:
      continue

remove_unused_words(toko_sort_rele['text'])
remove_unused_words(toko_sort_lowe['text'])
remove_unused_words(toko_sort_high['text'])
remove_unused_words(toko_filt_lowe['text'])
remove_unused_words(toko_filt_high['text'])

toko_sort_rele = toko_sort_rele[toko_sort_rele['text'].str.contains("deleted")==False]
toko_sort_lowe = toko_sort_lowe[toko_sort_lowe['text'].str.contains("deleted")==False]
toko_sort_high = toko_sort_high[toko_sort_high['text'].str.contains("deleted")==False]
toko_filt_lowe = toko_filt_lowe[toko_filt_lowe['text'].str.contains("deleted")==False]
toko_filt_high = toko_filt_high[toko_filt_high['text'].str.contains("deleted")==False]

toko_sort_rele.reset_index(inplace=True, drop=True)
toko_sort_lowe.reset_index(inplace=True, drop=True)
toko_sort_high.reset_index(inplace=True, drop=True)
toko_filt_lowe.reset_index(inplace=True, drop=True)
toko_filt_high.reset_index(inplace=True, drop=True)
```

```python
elec_1 = ('case')
elec_2 = ('redmi')
elec_3 = ('10a')
elec_4 = ('10 a')

def elec_chec_keyw(data_details):
  falseCounter=0
  for i in range (len(data_details)):
    if elec_1 in data_details[i]:
      if elec_2 in data_details[i]:
        if elec_3 in data_details[i]:
          continue
        if elec_4 in data_details[i]:
          continue
        else:
          falseCounter+=1
      else:
          falseCounter+=1
    else:
      falseCounter+=1
  print('accuracy of sort by relevance tokopedia: ', (len(data_details)-falseCounter)/len(data_details))
  print('true items, :', len(data_details)-falseCounter, ', false items :', falseCounter)
```
